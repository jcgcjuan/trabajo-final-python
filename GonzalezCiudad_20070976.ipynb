**HACIENDO SUBSETS**

1) Debido a que la base de datos es bien grande, y realmente lo que buscamos es hacer una comparación simple pero interesante, en primer lugar, vamos a quedarnos con las variables relacionadas directamente a población, condición económica y migración.

```python
nomb_variables = df2.columns
print(nomb_variables) #primero vemos los nombres de cada variable para saber con qué quedarnos
```

```python
#vamos eliminando las variables innecesarias
columnas = ["country", "region", "year", "population", "pop_density", "net_migration", "migration_perc", "incomeLevel" ]
migra = df2.loc[:, columnas]
migra
```

2) En segundo lugar, luego de haber escogido las variables importantes, buscamos también qué casos serán de nuestro interés. Para nuestra comparación, vamos a elegir los casos que pertenezcan a las regiones de America del Norte y Europa

```python
migra['region'].value_counts() #para saber qué categorías tenía la variable región
```

```python
migra_2 = migra.set_index(["region", "country"]).sort_index() #vamos a poner como index a la región, y los países. De esta manera luego podremos filtrar los casos en base a ello.
migra_2
```

```python
filas = ["Europe & Central Asia", "North America"] #utilizando el comando LOC, vamos a quedarnos con los casos que tengan como index la región de europa central asia y norte america.
lamigra= migra_2.loc[filas, :]
lamigra
```

Como los Censos son cada 5 años, se ha utilizado el comando DROPNA para eliminar los valores perdidos de los años que no hubo conteo.


```python
lamigra = lamigra.dropna()
lamigra
```

Ahora imaginemos que queremos hacer un análisis de cómo ha ido evolucionando la migración en los continentes seleccionados. 

Por tanto, buscaremos quedarnos con los casos de los países, según 7 años clave para ver la evolución: 1987, 1992, 1997, 2002, 2007, 2012, 2017

```python
lamigra = lamigra[lamigra.year.isin([1987, 1992, 1997, 2002, 2007, 2012, 2017])]
lamigra
```

```python
lamigra.info()
lamigra = lamigra.reset_index() #vamos a resetear los indices para poder hacer bien los gráficos
lamigra.info() #aquí ya tenemos las variables nuevamente como antes.
```

Ultimas modificaciones y limpiezas antes de hacer los gráficos

```python
#vamos eliminando las variables innecesarias
columnas = ["country", "region", "year", "population", "pop_density", "net_migration", "migration_perc", "incomeLevel" ]
lamigra = lamigra.loc[:, columnas]
lamigra
```

```python
#vamos cambiando el nombre de los casos de europoa y asia....ya que nos hemos quedado con solo algunos paises importantes de la UE
lamigra.loc[lamigra ["region"] == "Europe & Central Asia", "region"] = "Europa principal"
lamigra
```

**VISUALIZACIÓN DE DATOS**

Finalmente vamos a realizar la visualización de datos


```python
import altair as alt
```

**Visualizando el porcentaje de migración por paises**

x: años
y: migración porcentual
colores: países
leyenda: países

```python
alt.Chart(lamigra).mark_line().encode(
    x = alt.X("year:Q", scale=alt.Scale(domain=[1985, 2020])),
    y = alt.Y("migration_perc:Q"),
    color = alt.Color("country"),
    column = alt.Column("region"),
    tooltip = ["country", "migration_perc"]
    ).interactive()
```
